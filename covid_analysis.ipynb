```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the CSV file
df = pd.read_csv("covid_data.csv")

# Step 2: Show first 5 rows
print("Sample Data:\n", df.head())

# Step 3: Top 10 countries with highest confirmed cases
top10 = df.sort_values(by='Confirmed', ascending=False).head(10)

# Step 4: Plotting the top 10
plt.figure(figsize=(10,6))
plt.bar(top10['Country'], top10['Confirmed'], color='skyblue')
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")
plt.title("Top 10 Countries - COVID Confirmed Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
